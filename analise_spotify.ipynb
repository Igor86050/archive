```python
import pandas as pd
df = pd.read_csv("spotify_tracks.csv")
import numpy as np
print(df.head())
```

```python
#quantas musicas existem no dataset
df['name'].count()
```

```python
df["artists"].unique
```

```python
#quantas musicas diferentes existem no dataset
df['name'].nunique()
```

```python
generos_mais_ouvidos =df['genre'].value_counts().sort_values(ascending=False).head(10)
print(f'Top 10 generos mais ouvidos: {generos_mais_ouvidos}')
```

```python
artistas_mais_populares = df.groupby("artists")['popularity'].mean().sort_values(ascending=False).head(10)
print(f'top 10 artistas mais populares: {artistas_mais_populares}')
```

```python
top_10_musicas_mais_populares = df.groupby("name")['popularity'].mean().sort_values(ascending=False).head(10)
print(f'top 10 músicas mais populares: {top_10_musicas_mais_populares}')
```

```python
genero_mais_popular = df.groupby("genre")['popularity'].mean()
print(f'genero mais popular: {genero_mais_popular.idxmax()}')
```

```python
media_pop_music = df.groupby("name")['popularity'].mean()
print(f'media de popilaridade das musicas: {media_pop_music}')
```

```python
#musica mais popular por artista
df.loc[df.groupby("artists")["popularity"].idxmax()][["artists", "name", "popularity"]]
```

```python
#musica mais popular de cada artista
df.loc[df.groupby("artists")["popularity"].idxmax()][['artists', 'name', 'popularity']]
```

```python
duracao_media_por_musica =df['duration_ms'].mean()
print(f'duracao media por musica: {duracao_media_por_musica}')
```

```python
import matplotlib.pyplot as plt

top_10_artistas_mais_populares = (
    df.groupby("artists")["popularity"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))

plt.bar(
    top_10_artistas_mais_populares.index,
    top_10_artistas_mais_populares.values
)

plt.title("Top 10 artistas mais populares")
plt.xlabel("Artistas")
plt.ylabel("Popularidade média")

plt.xticks(rotation=45)

plt.show()
```

```python
artista_com_mais_musicas = df["artists"].value_counts().idxmax()

print(f'Artista com mais músicas: {artista_com_mais_musicas}')
```

```python
#top 10 artistas mais frequnetes
df["artists"].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

musicas_mais_populares = (
    df.sort_values(by="popularity", ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))

plt.bar(
    musicas_mais_populares["name"],
    musicas_mais_populares["popularity"]
)

plt.title("Top 10 músicas mais populares")
plt.xlabel("Músicas")
plt.ylabel("Popularidade")

plt.xticks(rotation=45)

plt.show()
```

```python
media_pop_music = df['popularity'].mean()

musicas_acima_media = df[df['popularity'] > media_pop_music]

print(musicas_acima_media)
```

```python
media_pop_por_musica = df.groupby("name")["popularity"].mean()

def classificar(pop):
    if pop < 30:
        return "baixa"
    elif pop < 50:
        return "moderada"
    else:
        return "alta"

categorias = media_pop_por_musica.apply(classificar)

print(categorias)
```
